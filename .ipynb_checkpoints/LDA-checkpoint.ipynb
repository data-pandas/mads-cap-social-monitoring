```python
# imports
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import warnings
warnings.filterwarnings("ignore")

#read csv
df = pd.read_csv('cleaned_non-vectorized_data.csv')
df = df[~df['Text'].isna()]
df
```

```python
import gensim.corpora as corpora

# Create Dictionary
id2word = corpora.Dictionary(df['Text'].str.split())
# Create Corpus
texts = df['Text'].str.split()
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
```

```python
from pprint import pprint
import gensim
# number of topics
num_topics = 5
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
```

```python
import re
for topic in lda_model.print_topics():
    i = topic[0]#topic number
    s = topic[1] #string of topics
    l=s.split("+") #list of terms
    terms=[]
    for term in l:
        stripped_term = (re.findall('".+"', term)[0].strip()[1:-1])
        terms.append(stripped_term)
    print("The top words in topic", i, "is", terms)
```
