```python
import pandas as pd
import glob
```

```python
#get all csv files in the labled csv folder 
file_names = glob.glob('Labeled csv\*.csv')
file_names
```

```python
list_of_df=[] #stores a list of df
for file in file_names:
    temp = pd.read_csv(file) #read csv
    list_of_df.append(temp) #add to list
    
df = pd.concat(list_of_df).loc[:,['Text','category','labels']] #concat all the dataframes into one 
df
```

```python
def string_to_list(s):
    '''
    turns the string "('n', 'n', 'n', 'n', 'n', 'n', 'y')" into a list [0,0,0,0,0,0,1]
    '''
    li = s[1:-1].split(',') #list of 'y' and 'n'
    li1=[] #new final list of 1 and 0
    for i in li:#iterate through y and n
        a=i.strip()[1] #extract the string 
        if a=='y':
            li1.append(1)
        else: #a=='n'
            li1.append(0)
    return li1

df['labels'] = df['labels'].apply(string_to_list) #apply the above function 
```

```python
#new df with the 1 and 0
df2 = pd.DataFrame(df['labels'].tolist(), index=df.index, columns=['disgust','joy','anger','surprised','sad','fear','neutral'])

df2['Text'] = df['Text'] #add the text column back in 
df_final = df2.loc[:,['Text','disgust','joy','anger','surprised','sad','fear','neutral']] #reorder the columns 

df_final
```

```python
df_final.to_csv('combined_labeled_data.csv')
```
