```python
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import string
import nltk

df = pd.read_csv('../../../Desktop/serve_data.csv') #read csv
df
```

```python
def text_preprocess(s):
    '''
    lowercase, lemmatization, and removal of stop words, removal of punctuation, removal of companies
    '''
    try:
        ind = s.index('http')
        s = s[:ind]
    except:
        pass 
    
    lem=WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    companies = set(['instacart','deliverydotcom','gopuff','grubhub','seamless','chownow','doordash','ubereats','uber eats','postmates','uber'])
    removal = stop_words.union(string.punctuation).union(companies)
    tokens = [lem.lemmatize(w.lower()) for w in nltk.word_tokenize(s) if not w.lower() in removal]
    return tokens

df['tokens'] = df.apply(lambda row: text_preprocess(row['text']), axis=1) #apply preprocessing to create tokens
df
```

```python
def join(li):
    return ' '.join(li)

df['text'] = df.apply(lambda row: join(row['tokens']), axis=1) #change Text to the preprocessed version. 
# df = df.loc[:,['text','tokens','disgust', 'joy', 'anger', 'surprised', 'sad', 'fear', 'neutral']] #reorder columns
df = df[~df['text'].isna()]
df
```

```python
df.to_csv('cleaned_non-vectorized_serve_data.csv', index=False)
```
